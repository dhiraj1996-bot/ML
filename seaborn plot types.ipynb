```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\csd\Desktop\ml\Python_ML_new-master\Automobile_data.csv")
```

```python
df
```

```python
x=np.arange(10)
y=np.sqrt(x)
plt.figure(figsize=(6,6))
sns.lineplot(x,y)
plt.show()
```

```python
#histogram
mean=df["width"].mean()
median=df["width"].median()
plt.figure(figsize=(6,6))
sns.distplot(df["width"])
plt.axvline(mean,color="black")
plt.axvline(median,color="yellow")
plt.show()
```

```python
#count plot
plt.figure(figsize=(6,6))
sns.countplot(data=df,x="body-style")
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(data=df,y="make")
#plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(data=df,x="body-style",hue="fuel-type")
plt.show()
```

```python
corr=df.corr()
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(corr,annot=True,cmap="viridis")
plt.show()
```

```python
#scatter plot(always done between continuos data vs the target)
plt.figure(figsize=(6,6))
sns.scatterplot(data=df,x="width",y="price")
plt.plot()
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(data=df,x="width",y="price",hue="body-style")
plt.plot()
```

```python
#boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x="price",y="make")
plt.plot()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(data=df,x="price",y="make")
plt.plot()
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(data=df,x="price",y="make")
plt.plot()
```

```python
fig, axs = plt.subplots(2, 2,figsize=(10,10))
sns.boxplot(data=df,x="price",y="make",ax=axs[0, 0])

sns.scatterplot(data=df,x="width",y="price",hue="price",ax=axs[0, 1])

sns.countplot(data=df,x="body-style",hue="fuel-type",ax=axs[1, 0])

sns.distplot(df["width"],ax=axs[1, 1])
plt.show()

```
