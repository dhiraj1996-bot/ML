```python
'''
->Regularization has two types or methods(1.)Lasso and (2.)Ridge
->while plotting the training data gives zero error but the testing data gives errors..
->at this point we use regularization methods --Lasso-- and --Ridge--..
->here we increase the error of the over fit the model so as to get a proper model
->there is lambda(hyperparameter) constant that can be between 0 to infinity
->in Lasso we take the absolute value(theta)
->in Ridge we take square of the absolute value(theta)
->Ridge decrease the value of theta but never makes it zero
->Lasso decreases the value of theta and if necessary makes it zero
->both Ridge and Lasso decreases only the features which are not that important
'''
```

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
for i in range(50):
    l2=Ridge(alpha=i)
    l2.fit(x_train,y_train)
    print(i,":",l2.score(x_test,y_test))
```

```python
for i in range(500):
    l1=Lasso(alpha=i)
    l1.fit(x_train,y_train)
    print(i,":",l1.score(x_test,y_test))
```

```python
l2=Ridge(alpha=3)
l2.fit(x_train,y_train)
print(l2.score(x_test,y_test))
```
