#DROPING NULL VALUES....

```python
name=["john","jack","jill","kenny","kevin"]
math=[89,78,67,np.nan,85]
science=[78,75,np.nan,np.nan,76]
```

```python
dic={"name":name,"math":math,"science":science}
```

```python
df=pd.DataFrame(dic)
```

```python
df
```

```python
#this will give the number of null values in the dataframe.
df.isnull().sum()
```

```python
#to drop the null values in the dataframe.
df.dropna()
```

```python
#when you just need to drop the particular row having null values.
df.dropna(axis=1)
```

```python
df.dropna(thresh=2)
#this will delete the row which has more than two or equal o two null values.
```

```python
df.dropna(subset=["name","math"])
```

```python
df.dropna(subset=["name","science"])
```

```python
#this will drop any row or column randomly.
df.drop(3)
```

```python
df.drop([1,3])
```

```python
df.drop(["name","math"],axis=1)
```

### REPLACING MISSING VALUES 

```python
df
```

```python
df.fillna('-')
#this will replace the nan values with blank or zero.
#to save the changes use "inplace=True".
```

```python
df.fillna(0)
```

```python
df.fillna(method="ffill")
#this function will fill the null area with the value before it.
#this method is not used much in real life.
```

```python
math_mean=df["math"].mean()
math_mean=np.ceil(math_mean)
science_mean=df["science"].mean()
science_mean=np.ceil(science_mean)
print(math_mean)
print(science_mean)
```

```python
df["math"].fillna(math_mean,inplace=True)
```

```python
df["science"].fillna(science_mean,inplace=True)
```

```python
df
```
